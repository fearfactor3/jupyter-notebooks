# Sorting Jira Issues by Categories
In this notebook, we will load the CSV file containing Jira issues, and then sort the issues by categories such as Issue Type, Priority, and Status.

```python
import pandas as pd

# Load the Jira issues CSV file
file_path = 'sample_jira_issues.csv'  # Update with the correct file path
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

### Sorting Issues by Type

```python
# Sort Jira issues by 'Issue_Type'
issues_by_type = df.sort_values(by='Issue Type')
issues_by_type.head()
```

### Sorting Issues by Priority

```python
# Sort Jira issues by 'Priority'
issues_by_priority = df.sort_values(by='Priority')
issues_by_priority.head()
```

### Sorting Issues by Status

```python
# Sort Jira issues by 'Status'
issues_by_status = df.sort_values(by='Status')
issues_by_status.head()
```

### Export results to excel spreadsheet for use outside of this.

```python
with pd.ExcelWriter('sorted_jira_issues_by_status.xlsx', engine='xlsxwriter') as writer:
    issues_by_priority.to_excel(writer, sheet_name='Sorted by Priority', index=False)
    issues_by_status.to_excel(writer, sheet_name='Sorted by Status', index=False)
    issues_by_type.to_excel(writer, sheet_name='Sorted by Type', index=False)
```
